# Julia Sets

These plots are created by out julia() function applied to the complex plane. This generates interesting fractal patterns that change as the complex parameter c is changed.

```python
c = 0
cp.JuliaPlane(-2,2,-2,2, c).show()
```

```python
c = -1
cp.JuliaPlane(-2,2,-2,2, c).show()
```

```python
c = 0.3
cp.JuliaPlane(-2,2,-2,2, c).show()
```

```python
c = -1.037 + 0.17j
cp.JuliaPlane(-2,2,-2,2, c).show()
```

```python
c = -0.624 + 0.435j
cp.JuliaPlane(-2,2,-2,2, c).show()
```

```python
c = -0.8 + 0.2j
cp.JuliaPlane(-2,2,-2,2, c).show()
```

```python
c = -0.835 - 0.2321j
cp.JuliaPlane(-2,2,-2,2, c).show()
```

# Changing color maps

For fun, I added the ability to change the color map being used.

```python
r = 0.75
theta = np.pi*1.45
c = r*(np.cos(theta) + np.sin(theta)*1j)
cp.JuliaPlane(-1.3,-1.2,-0.6,-0.5, c, 1000).show('prism')
```
